```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return x[0] ** 2 + x[1] ** 2 - 4

def f2(x):
    return x[0] - np.exp(x[1]) + 2

def df1_x1(x):
    return 2 * x[0]

def df1_x2(x):
    return 2 * x[1]

def df2_x1(x):
    return 1

def df2_x2(x):
    return - math.exp(x[1])

a = -1.5
b = 3.5
```

```python
x = np.arange(a, b, 0.01)
y = np.arange(a, b, 0.01)
x, y = np.meshgrid(x,y)
z1 = x ** 2 + y ** 2 - 4
z2 = x - np.exp(y) + 2
plt.contour(x,y,z1,levels =0,colors = 'green' )
plt.contour(x,y,z2,levels =0, colors = 'blue')
```

```python
x = np.arange(1.2,2,0.01)
y = np.arange(1.1,1.4,0.01)
x, y = np.meshgrid(x,y)
z1 = x ** 2 + y ** 2 - 4
z2 = x - np.exp(y) + 2
plt.contour(x,y,z1,levels =0,colors = 'green' )
plt.contour(x,y,z2,levels =0, colors = 'blue')
```

положительное решение находится в 1.5 < x1 < 1.6, 1.25 < x2 < 1.30


Метод Ньютона¶


```python
def A1(x):
    a = np.zeros((2,2), dtype=int)
    a[0, 0] = f1(x)
    a[1, 0] = f2(x)
    a[0, 1] = df1_x2(x)
    a[1, 1] = df2_x2(x)
    return np.linalg.det(a)

def A2(x):
    a = np.zeros((2,2), dtype=int)
    a[0, 0] = df1_x1(x)
    a[1, 0] = df2_x1(x)
    a[0, 1] = f1(x)
    a[1, 1] = f2(x)
    return np.linalg.det(a)

def J(x):
    a = np.zeros((2,2), dtype=int)
    a[0, 0] = df1_x1(x)
    a[1, 0] = df2_x1(x)
    a[0, 1] = df1_x2(x)
    a[1, 1] = df2_x2(x)
    return np.linalg.det(a)
```

```python
def newton_next(x):
    y = [0, 0]
    y[0] = x[0] - A1(x) / J(x)
    y[1] = x[1] - A2(x) / J(x)
    return y


def newton(a, b, eps):
    count = 0
    x = [a, b]
    x_next = newton_next(x)
    while max(abs(x[0] - x_next[0]), abs(x[1] - x_next[1])) > eps:
        x = x_next
        x_next = newton_next(x)
        count += 1
    return x_next, count
```

```python
a1 = 0.25
b1 = 0.75
newton(a1, b1, 0.001)
```

```python
# зависимость от кол-ва итераций
point = {i : [newton(a1, b1, i * 0.01)[1], i * 0.01] for i in range(100) }
for i in point:
    plt.scatter(*point[i])
plt.xlim(0, 10)
plt.ylim(0, 1.3)
plt.ylabel('eps')
plt.show()
```

```python
def phi1(x):
    return np.sqrt(4 - x[0] ** 2)
    
def phi2(x):
    return np.exp(x[1]) - 2

def dphi1_x1(x):
    return - x[0] / (4 - x[0] ** 2)

def dphi1_x2(x):
    return 0

def dphi2_x1(x):
    return 0

def dphi2_x2(x):
    return np.exp(x[1])

def find_q(x):
    maxphi1 = dphi1_x1(x) + dphi1_x2(x)
    maxphi2 = dphi2_x1(x) + dphi2_x2(x)
    return max(maxphi1, maxphi2)
```

```python
a1 = 0.5
b1 = 0.2
x = [a1, b1]
find_q(x)
```

```python
def iteration_next(x):
    y = [0, 0]
    y[0] = phi1(x)
    y[1] = phi2(x)
    return y

def simple_iter(a, b, eps):
    count = 0
    x = [a, b]
    q = find_q(x)
    x_next = iteration_next(x)
    while max([abs(i - j) for i, j in zip(x_next, x)]) * q / (1 - q) > eps:
        x = x_next
        x_next = iteration_next(x)
        count += 1
    return x_next, count
```

```python
a1 = 1.5
b1 = 1.001
simple_iter(a1, b1, 0.00001)
```

```python
# зависимость от кл-ва итераций
point = {i : [simple_iter(a1, b1, i * 0.01)[1], i * 0.01] for i in range(100) }
for i in point:
    plt.scatter(*point[i] )
plt.xlim(-2, 2)
plt.ylim(0, 1)
plt.ylabel('eps')
plt.show()
```
